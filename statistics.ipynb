```python
# 绘制折线图
def draw_line_chart():
    data = pd.read_csv('res/results.csv', delimiter=',')
    print('shape = ', data.shape)
    dt_avg = np.array(data['average'])
    dt_var = np.array(data['var'])
    dt_std = np.array(data['std'])
    avg = np.average(dt_avg)
    print('avg = ', avg)
    print('均值: \n', dt_avg)
    print('方差: \n', dt_var)
    print('标准差:\n', dt_std)

    print(data.values)

    all_std = np.std(data.values)
    # all_data = data.values()[:, :]
    # print(all_data)
    # print('整体标准差 = ', np.std(np.array(all_data)))
    # 均值，方差，标准差的折线图
    x = np.arange(1, data.shape[0] + 1)
    # np.set_printoptions(precision=3)
    # y = np.linspace(0, 2.8, 15)
    #
    # y = np.round(y, decimals=2)
    print(x)
    # print(y)
    plt.figure(figsize=(10, 6))
#     plt.figure(figsize=(12, 8))
#     plt.tick_params(labelsize=16)
    plt.plot(x, dt_avg, 'r-', linewidth=1, label='Average of relative errors in one trail')
    plt.plot(x, np.array([avg] * len(dt_avg)), 'r--', linewidth=1, label='Average of relative errors in all trails')
    plt.plot(x, np.array([all_std]*data.shape[0]), 'm--', linewidth=1, label='Standard deviation of relative errors in all trails')
    # plt.plot(x, dt_var, 'b-', linewidth=2, label='Var')
    plt.text(100, avg - 0.05, '1.40')
    plt.text(100, all_std - 0.05, round(all_std, 2))
#     plt.text(100, avg - 0.05, round(avg, 3), fontsize=16)
#     plt.text(100, all_std - 0.05, round(all_std, 3), fontsize=16)

    plt.ylim(0, 5)
    plt.xlim(1, data.shape[0])
    plt.xlabel('Trails')
    plt.ylabel('Relative errors between two blocks (mm)')
    plt.legend(loc='upper right')
#     plt.xlabel('Trails', fontsize=16)
#     plt.ylabel('Relative errors between two blocks (mm)', fontsize=16)
#     plt.legend(loc='upper right', fontsize=16)
    # plt.title('Results')
    # plt.grid()
    plt.savefig('res/statistics.jpg')
    plt.savefig('res/statistics.svg')
    plt.show()
    
if __name__ == '__main__':
    draw_line_chart()
```

```python
# 绘制频方图
def draw_hist_2():
    data = pd.read_csv('res/results.csv', delimiter=',')
    plt.figure(figsize=(10, 6))
    n, bins, patches = plt.hist(data['average'], bins=15, facecolor='blue', alpha=0.75, rwidth=0.8)
    # plt.hist(data['average'], bins=15, normed=1, facecolor='blue', alpha=0.6, rwidth=0.8, cumulative=True)
    plt.plot(np.array([1.4] * 16) , np.arange(0, 16), 'r--', label='Average of relative errors in all trails')
    # plt.grid()
    y_list = n
    x_list = bins[: -1]
    print('y_list = ', y_list)
    for i, x in enumerate(x_list):
        if y_list[i] != 0:
            if y_list[i] > 9:
                plt.text(x + 0.049, y_list[i] + 0.2, np.int(y_list[i]))
            else:
                plt.text(x + 0.065, y_list[i] + 0.2, np.int(y_list[i]))
    
    plt.text(1.4 - 0.05, 0 - 0.83, 1.4)
    
    print('n = ', n)
    print('bins = ', bins)
    print('patches = ', patches)
    print('len(n) = ', len(n))
    print('len(bins) = ', len(bins))
    plt.legend(loc='upper right')
    plt.ylim(0, 18, 2)
    plt.xlabel('Relative errors between two blocks (mm)')
    plt.ylabel('Trail times')

    plt.savefig('res/hist2.jpg')
    plt.savefig('res/hist2.svg')

    plt.show()

draw_hist_2()
```
